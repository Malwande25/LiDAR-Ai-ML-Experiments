```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Loading the logged CSV file
df = pd.read_csv("sf45b_log.csv")

# Quick check
print(df.head())

# Extract columns
yaw_angles = df['yaw_deg'].values   # in degrees
distances = df['distance_m'].values  # in meters

# Convert yaw to radians
yaw_rad = np.deg2rad(yaw_angles)

# Convert polar -> Cartesian
x = distances * np.cos(yaw_rad)
y = distances * np.sin(yaw_rad)

# Plot scan
plt.figure(figsize=(7,7))
plt.plot(x, y, 'bo-', label="Scan shape")
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.title("LiDAR Scan Shape from CSV Data")
plt.legend()
plt.axis("equal")
plt.grid(True)
plt.show()

# Example plot
#plt.figure(figsize=(10,5))
#plt.scatter(df["yaw_deg"], df["distance_m"], s=10)
#plt.xlabel("Yaw (degrees)")
#plt.ylabel("Distance (m)")
#plt.title("LiDAR Scan")
#plt.show()
```
